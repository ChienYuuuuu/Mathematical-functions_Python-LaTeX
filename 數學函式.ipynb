```python
import matplotlib.pyplot as plt

def calculate_Sn(n):
    Sn = [sum(1/k for k in range(1, i+1)) for i in range(1, n+1)]
    return Sn

n_values = range(1, 11)  # 選擇你要顯示的範圍，這裡是 1 到 10

Sn_values = calculate_Sn(max(n_values))

plt.plot(n_values, Sn_values, marker='o')
plt.xlabel('n')
plt.ylabel('S_n')
plt.title(r'$S_n=\sum_{k=1}^{n} \frac{1}{k}$')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n=100
x = np.linspace(1, 100, 200) 
f = 1
g = 1/2(1-1/(n+1))

def calculate_sum(n):
    total = 0
    for k in range(1, n+1):
        total += 1/k
    return total
h = calculate_sum(n)-ln(n+1)

plt.plot(f)
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.grid(1)
#plt.xlabel('X')
#plt.ylabel('f(x)', rotation=0)
#plt.title(r'$f(x)= \frac{1}{x}$')
plt.show() 

```

```python
def f(n):
    return 0.5 * (1 - (1 / (n + 1)))

def S_n(n):
    return sum(1/i for i in range(1, n+1))

def gamma_n(n):
    return S_n(n) - np.log(n + 1)

n_values = np.arange(0, 101)

# 計算 f(n) 和 gamma_n(n) 的值
f_values = [f(n) for n in n_values]
gamma_values = [gamma_n(n) for n in n_values]
plt.figure(figsize=(8, 6))
plt.axhline(y = 1, color='red', label = r'1')
plt.plot(n_values, gamma_values, linestyle='-', color = 'orange', label = r'$\gamma_n$')
plt.plot(n_values, f_values, linestyle='-', color = 'green', label = r'$\frac{1}{2} (1 - \frac{1}{n + 1})$')

plt.xlabel('n')

plt.legend()
plt.grid(True)
plt.savefig('poly_25.eps', format='eps')
plt.show()
```
